```r
library('tidyverse')
```

```r
data <- read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-10-27/wind-turbine.csv")
```

Each row in the table represents a single turbine.

Make a bar chart 沒 **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
data1 <- data |> mutate(province_territory = province_territory |> fct_infreq() |> fct_rev())

p <- ggplot(data1, aes(y = province_territory)) + geom_bar(fill = "forest green") + labs(y = "Province/Territory", x = "Number of Turbines")

p
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
data_man <- data |> mutate(manufacturer = manufacturer |> fct_lump_n(5) |> fct_infreq() |> fct_rev())

p1 <- ggplot(data_man, aes(y = manufacturer)) + geom_bar(fill = "forest green") + labs(y = "Manufacturer", x = "Number of Turbines made")

p1
```
